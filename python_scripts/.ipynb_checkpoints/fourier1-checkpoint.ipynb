```python
from matplotlib import pyplot as plt
import numpy as np
import scipy as sp
from scipy.io import wavfile
from scipy.fftpack import fft, ifft
import IPython
```

# Fourier Analysis
Code to clean audio files using Fourier transformation, as well as code to create sounds using f

```python
#problem 1
class Signal(object):
    
    def __init__(self, rate, sample):
        """This makes the code light on fire.
        rate: rate of the sample
        sample: array of samples"""
        self.rate=rate
        self.sample=sample
        
        #problem 5
    def plot(self):
        """This makes Koa light on fire."""
        x=np.linspace(0,len(self.sample)/self.rate, len(self.sample))
        plt.subplot(121)
        plt.plot(x, self.sample)
        dft=sp.fft(self.sample)
        x_vals=sp.arange(1, len(dft)+1, 1)*1.
        x_vals=x_vals/len(self.sample)
        xvals=x_vals*rate
        plt.subplot(122)
        plt.plot(x_vals, dft)
        plt.show()
        
    #problem 2    
    def export(self, filename):
        samples=sp.int16(self.sample*32767./np.max(np.abs(self.sample)))
        wavfile.write(filename, self.rate, samples)
        
    def __add__(self, other):
        sampled=other.sample+self.sample
        return Signal(self.rate, sampled)
```

```python

rate, wave=wavfile.read('tada.wav')
```

```python
sig=Signal(rate, wave)
```

```python
%matplotlib inline
sig.plot()
sig.export('text.wav')
```

```python
#problem 3. Play an A

def note_generator(samplerate, frequency):
    stepsize=1./samplerate
    sample_points=sp.arange(0, 5,stepsize)
    wave_function=lambda x:sp.sin(2*sp.pi*x*frequency)
    samples=wave_function(sample_points)
    return Signal(samplerate, samples)

samplerate=44100
a_frequency=440
a_wave=note_generator(samplerate, a_frequency)
a_wave.plot()
a_wave.export("a.wav")
IPython.display.Audio("a.wav")
```

```python
#problem 4

def discrete_fourier(array):
    l=[]
    for i in array:
        c=0
        for j in array[i]:
            c=c+array[i,j]*e**(2*np.pi*np.i*j*i/len(array))
        l.append(c)
    return l
```

```python
#problem 6
c_frequency=523.25
e_frequency=659.25
f_frequency=698.46
d_frequency=587.33
g_frequency=783.99
C=note_generator(samplerate, c_frequency)
E=note_generator(samplerate, e_frequency)
A=note_generator(samplerate, a_frequency)
F=note_generator(samplerate, f_frequency)
D=note_generator(samplerate, d_frequency)
G=note_generator(samplerate, g_frequency)
chord=A+C+E
chord1=A+C+F
chord2=A+D+F
chord3=C+E+G
chord.plot()
chord1.plot()
modulation=np.hstack((chord.sample, chord1.sample, chord2.sample, chord3.sample))
modulation=Signal(samplerate, modulation)
modulation.export("mod.wav")
IPython.display.Audio("mod.wav")
```
